```python
from PyPDF2 import PdfReader

try:
    with open('Business_Proposal.pdf', 'rb') as file:
        reader = PdfReader(file)
        page_two_text = reader.pages[1].extract_text()
        print("Text from page 2 extracted successfully.")
except FileNotFoundError:
    print("Error: Business_Proposal.pdf not found.")
```

```python
try:
    with open('contacts.txt', 'a') as file:
        file.write(page_two_text)
        print("Text from page 2 appended to contacts.txt.")
except Exception as e:
    print(f"Error appending to file: {e}")
```

```python
import re

email_pattern = r'\S+@\S+'
email_addresses = re.findall(email_pattern, page_two_text)

if email_addresses:
    print("Email addresses found:")
    for email in email_addresses:
        print(email)
else:
    print("No email addresses found.")
```
